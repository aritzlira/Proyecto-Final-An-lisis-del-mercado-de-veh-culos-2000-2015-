```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../datos/procesados/dataset_final_coches.csv")
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["sellingprice"].describe()
```

```python
plt.hist(df["sellingprice"], bins=50)
plt.title("Distribución del precio de venta")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df["year"].value_counts().sort_index().plot(
    kind="bar", figsize=(12,4)
)
plt.title("Distribución de vehículos por año")
plt.show()
```

```python
df["make"].value_counts().head(10)
```

```python
df["make"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 marcas por número de registros")
plt.show()
```

```python
df["fuel consumption"].isnull().mean()
```

```python
df_consumo = df.dropna(subset=["fuel consumption"])
```

```python
plt.hist(df_consumo["fuel consumption"], bins=50)
plt.title("Distribución del consumo de combustible")
plt.xlabel("Consumo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.scatter(
    df_consumo["fuel consumption"],
    df_consumo["sellingprice"],
    alpha=0.2
)
plt.xlabel("Consumo de combustible (L/100 km)")
plt.ylabel("Precio de venta")
plt.title("Relación entre precio y consumo de combustible")
plt.show()
```
