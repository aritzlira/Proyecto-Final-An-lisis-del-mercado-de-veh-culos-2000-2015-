```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../datos/procesados/dataset_final_coches.csv")
df.shape
```

```python
vars_num = [
    "sellingprice",
    "fuel consumption",
    "odometer",
    "engine size",
    "cylinders",
    "year"
]

df_stats = df[vars_num]
df_stats.info()
```

```python
df_stats_clean = df_stats.dropna()
df_stats_clean.shape
```

```python
corr_matrix = df_stats_clean.corr()
corr_matrix
```

```python
corr_matrix["sellingprice"].sort_values(ascending=False)
```

```python
## Análisis de correlaciones

El análisis de correlación muestra que el precio de venta de los vehículos está fuertemente relacionado con el año de fabricación, lo que indica que los vehículos más recientes tienden a venderse a precios más elevados. Asimismo, se observa una correlación negativa significativa entre el precio y el kilometraje, reflejando el efecto de la depreciación por uso.

Las variables técnicas como el tamaño del motor y el número de cilindros presentan correlaciones positivas moderadas con el precio, mientras que el consumo de combustible muestra una relación débil. Esto confirma que el precio de los vehículos está influenciado por múltiples factores y no depende únicamente del consumo.
```

```python
df_stats_clean.groupby("cylinders")["sellingprice"].mean()
```

```python
df_stats_clean.groupby("year")["sellingprice"].mean().tail(10)
```

```python
## Conclusiones del análisis estadístico

El análisis estadístico confirma cuantitativamente las tendencias observadas durante el análisis descriptivo. El precio de los vehículos está principalmente influenciado por el año de fabricación y el kilometraje, mientras que las características técnicas del motor aportan una influencia adicional. Por el contrario, el consumo de combustible presenta una relación débil con el precio. Estos resultados proporcionan una base sólida para el desarrollo del dashboard y del informe final.
```
