```python
import pandas as pd
import numpy as np
```

```python
df_prices = pd.read_csv("../datos/crudos/car_prices.csv")
df_fuel = pd.read_csv("../datos/crudos/Fuel_Consumption_2000-2022.csv")
```

```python
df_prices.columns = df_prices.columns.str.lower().str.strip()
```

```python
cols_numeric = ["year", "sellingprice", "mmr", "odometer"]
for col in cols_numeric:
    df_prices[col] = pd.to_numeric(df_prices[col], errors="coerce")
```

```python
df_prices["make"] = df_prices["make"].str.upper().str.strip()
df_prices["model"] = df_prices["model"].str.upper().str.strip()
```

```python
df_prices.isnull().sum()
df_prices.duplicated().sum()
```

```python
df_fuel.columns = df_fuel.columns.str.lower().str.strip()
```

```python
df_fuel = df_fuel.rename(columns={
    "year": "year",
    "make": "make",
    "model": "model"
})
```

```python
df_fuel["year"] = pd.to_numeric(df_fuel["year"], errors="coerce")
```

```python
df_fuel["make"] = df_fuel["make"].str.upper().str.strip()
df_fuel["model"] = df_fuel["model"].str.upper().str.strip()
```

```python
df_fuel.isnull().sum()
df_fuel.duplicated().sum()
```
